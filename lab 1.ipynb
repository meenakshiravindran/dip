```python
import cv2
import matplotlib.pyplot as plt

img=cv2.imread("parrot.jpeg")
neg_img=cv2.bitwise_not(img)
neg2_img=255-img
rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
rgb_neg=cv2.bitwise_not(rgb)
gray_neg=cv2.bitwise_not(gray)

plt.figure(figsize=(12,6))

plt.subplot(331)
plt.imshow(img)
plt.title("Image")
plt.axis("off")

plt.subplot(332)
plt.imshow(neg_img)
plt.title("Image")
plt.axis("off")

plt.subplot(333)
plt.imshow(neg2_img)
plt.title("Image")
plt.axis("off")

plt.subplot(334)
plt.imshow(rgb)
plt.title("Image")
plt.axis("off")

plt.subplot(335)
plt.imshow(rgb_neg)
plt.title("Image")
plt.axis("off")

plt.subplot(336)
plt.imshow(gray,cmap="gray")
plt.title("Image")
plt.axis("off")

plt.subplot(337)
plt.imshow(gray_neg,cmap="gray")
plt.title("Image")
plt.axis("off")

r,g,b = cv2.split(img)
print([r,g,b])

plt.imshow(r, cmap='gray'); plt.title("Red Channel"); plt.show()
plt.imshow(g, cmap='gray'); plt.title("Green Channel"); plt.show()
plt.imshow(b, cmap='gray'); plt.title("Blue Channel"); plt.show()
```

```python
threshold=127

_,binary = cv2.threshold(gray,threshold,255,cv2.THRESH_BINARY)
plt.imshow(binary, cmap='gray')
plt.title("Binary Image")
plt.axis('off')
```
