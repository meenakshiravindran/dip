```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np
from PIL import Image

im1=Image.open("dog.jpeg")
im2=Image.open("parrot.jpeg")

plt.figure(figsize=(12,6))

plt.subplot(221)
plt.imshow(im1,cmap="gray")
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(222)
plt.imshow(im2,cmap="gray")
plt.title("  Image 2")
plt.axis("off")

img1=np.array(im1)
img2=np.array(im2)

img1=cv2.resize(img1,(img2.shape[1],img2.shape[0]))

plt.subplot(223)
plt.imshow(img1,cmap="gray")
plt.title("Image 1 after resize")
plt.axis("off")

plt.subplot(224)
plt.imshow(img2,cmap="gray")
plt.title("  Image 2")
plt.axis("off")


mean_img1=np.mean(img1)
std_img1=np.std(img1)

mean_img2=np.mean(img2)
std_img2=np.std(img2)
corr=np.corrcoef(img1.flatten(),img2.flatten())[0,1]

print(mean_img1)
print(std_img1)

print(mean_img2)
print(std_img2)


print(corr)
```
