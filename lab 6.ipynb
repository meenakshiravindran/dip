```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img=cv2.imread("parrot.jpeg",cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error")
    exit()

threshold=100
_,thresh_img=cv2.threshold(img,threshold,255,cv2.THRESH_BINARY)

min_val=np.min(img)
max_val=np.max(img)
contrast_img=((img-min_val)/(max_val-min_val))*255
contrast_img=contrast_img.astype(np.uint8)

c=255/np.log(1+np.max(img))
log_float=c*np.log(1+img.astype(np.float32))
log_img=log_float.astype(np.uint8)


inv_log_img=np.exp(log_float/c) - 1
inv_log_img=np.clip(inv_log_img,0,255).astype(np.uint8)

gamma=0.5
normalized=img/255.0
gamma_corrected=np.power(normalized,gamma)
power_img=np.uint8(gamma_corrected*255)


plt.figure(figsize=(12,6))
plt.subplot(321)
plt.imshow(img,cmap="gray")
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(322)
plt.imshow(thresh_img,cmap="gray")
plt.title("Histogram Equalized Color")
plt.axis("off")

plt.subplot(323)
plt.imshow(contrast_img,cmap="gray")
plt.title("Histogram Equalized Color")
plt.axis("off")

plt.subplot(324)
plt.imshow(inv_log_img,cmap="gray")
plt.title("Histogram Equalized Color")
plt.axis("off")

plt.subplot(325)
plt.imshow(power_img,cmap="gray")
plt.title("Histogram Equalized Color")
plt.axis("off")

plt.subplot(326)
plt.imshow(log_img,cmap="gray")
plt.title("Histogram Equalized Color")
plt.axis("off")

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img=cv2.imread('dog.jpeg',cv2.IMREAD_GRAYSCALE)
if img is None:
    print("Error:Image not found or path is incorrect")
    exit()
threshold=100
_,thresh_img=cv2.threshold(img,threshold,255,cv2.THRESH_BINARY)
min_val=np.min(img)
max_val=np.max(img)
contrast_img=((img-min_val)/(max_val-min_val))*255
contrast_img=contrast_img.astype(np.uint8)
c=255/np.log(1+np.max(img))
log_img=c*np.log(1+img.astype(np.float32))
log_img=np.uint8(log_img)
iv_log_img=np.exp(log_img/c)
iv_log_img=np.clip(iv_log_img,0,255)
iv_log_img=np.uint8(iv_log_img)
gamma=0.5
normalized=img/255.0
gamma_corrected=np.power(normalized,gamma)
power_img=np.uint8(gamma_corrected*255)
plt.figure(figsize=(10,10))
plt.subplot(2,3,1)
plt.imshow(img,cmap='gray')
plt.title('orginal image')
plt.axis('off')
plt.subplot(2,3,2)
plt.imshow(thresh_img,cmap='gray')
plt.title('Threshold image')
plt.axis('off')
plt.subplot(2,3,3)
plt.imshow(contrast_img,cmap='gray')
plt.title('contrast image')
plt.axis('off')
plt.subplot(2,3,4)
plt.imshow(log_img,cmap='gray')
plt.title('logarithemic image')
plt.axis('off')
plt.subplot(2,3,5)
plt.imshow(power_img,cmap='gray')
plt.title('power image')
plt.axis('off') 
plt.subplot(2,3,6)
plt.imshow(iv_log_img,cmap='gray')
plt.title('Inverse image')
plt.axis('off')
```
