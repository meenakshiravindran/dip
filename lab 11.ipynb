```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img=cv2.imread("parrot.jpeg")
gblur=cv2.GaussianBlur(img,(3,3),0)
kernel=np.array([[0,-1,0],
                 [-1,5,-1],
                 [0,-1,0]])
sharpen_image=cv2.filter2D(img,-1,kernel)
edges=cv2.Canny(gblur,threshold1=100,threshold2=200)

plt.figure(figsize=(12,6))

plt.subplot(221)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(222)
plt.imshow(cv2.cvtColor(gblur,cv2.COLOR_BGR2RGB))
plt.title("Blurred Image")
plt.axis("off")

plt.subplot(223)
plt.imshow(cv2.cvtColor(edges,cv2.COLOR_BGR2RGB))
plt.title("edges of Image")
plt.axis("off")
```
