```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("parrot.jpeg",0)
min_intensity=np.min(img)
max_intensity=np.max(img)
streched_img=(((img-min_intensity)/(max_intensity-min_intensity))*255).astype(np.uint8)

hist_original=cv2.calcHist([img],[0],None,[255],[0,255])
hist_streched=cv2.calcHist([streched_img],[0],None,[255],[0,255])

equalized_img=cv2.equalizeHist(img)
hist_equalized=cv2.calcHist([equalized_img],[0],None,[255],[0,255])

plt.figure(figsize=(12,6))

plt.subplot(231)
plt.imshow(img,"Accent")
plt.title("Original Image")
plt.axis('off')

plt.subplot(232)
plt.imshow(streched_img,"Accent")
plt.title("Original Image")
plt.axis('off')

plt.subplot(233)
plt.imshow(equalized_img,"Accent")
plt.title("Original Image")
plt.axis('off')

plt.subplot(234)
plt.plot(hist_original)
plt.title("Original Image")
plt.xlim([0,256])

plt.subplot(235)
plt.plot(hist_streched)
plt.title("Original Image")
plt.xlim([0,256])

plt.subplot(236)
plt.plot(hist_equalized)
plt.title("Original Image")
plt.xlim([0,256])
```

```python
plt.figure()
plt.plot(hist_original, label="Original")
plt.plot(hist_streched, label="Stretched")
plt.plot(hist_equalized, label="Equalized")
plt.legend()
plt.title("Histogram Comparison")
plt.xlim([0,256])
plt.show()
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.figure(figsize=(10,4))
plt.bar(np.arange(256), hist[:,0])
plt.title("Histogram using plt.bar()")
plt.show()
```

```python
roi = img[50:200, 100:300]          # adjust these values based on your image

min_i = np.min(roi)
max_i = np.max(roi)

stretched_roi = (((roi - min_i) / (max_i - min_i)) * 255).astype(np.uint8)

# Replace ROI back into original image
img_roi_stretched = img.copy()
img_roi_stretched[50:200, 100:300] = stretched_roi

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.imshow(img, cmap="gray")
plt.title("Original Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(img_roi_stretched, cmap="gray")
plt.title("ROI Contrast Stretched")
plt.axis("off")
plt.show()
```

```python

color_img = cv2.imread("parrot.jpeg")
color_img = cv2.cvtColor(color_img, cv2.COLOR_BGR2RGB)

# Equalize each channel separately
r, g, b = cv2.split(color_img)
r_eq = cv2.equalizeHist(r)
g_eq = cv2.equalizeHist(g)
b_eq = cv2.equalizeHist(b)

equalized_color = cv2.merge((r_eq, g_eq, b_eq))

plt.figure(figsize=(10,4))
plt.subplot(121)
plt.imshow(color_img)
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(122)
plt.imshow(equalized_color)
plt.title("Histogram Equalized Color")
plt.axis("off")
plt.show()
```
