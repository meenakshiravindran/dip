```python
import cv2
import matplotlib.pyplot as plt 
import numpy as np

img=cv2.imread("parrot.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

def mean(img,ksize=3):
    h,w,c=img.shape
    k=ksize//2
    output=img.copy()
    for i in range(k,h-k):
        for j in range(k,w-k):
            for ch in range(c):
                region=img[i-k:i+k+1,j-k:j+k+1,ch]
                output[i,j,ch]=np.mean(region)
    return output

def median(img,ksize=3):
    h,w,c=img.shape
    k=ksize//2
    output=img.copy()
    for i in range(k,h-k):
        for j in range(k,w-k):
            for ch in range(c):
                region=img[i-k:i+k+1,j-k:j+k+1,ch]
                output[i,j,ch]=np.median(region)
    return output

mean=mean(img,ksize=3)
median=median(img,ksize=3)

plt.figure(figsize=(12,6))

plt.subplot(221)
plt.imshow(img)
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(222)
plt.imshow(mean)
plt.title("Original Color Image")
plt.axis("off")

plt.subplot(223)
plt.imshow(median)
plt.title("Original Color Image")
plt.axis("off")
```
