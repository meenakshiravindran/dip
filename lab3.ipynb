```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img=cv2.imread("parrot.jpeg")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

def resize(img,width,height):
    src_h,src_w=img.shape[:2]
    dst=np.zeros((height,width,3),dtype=np.uint8)
    for i in range(height):
        for j in range(width):
            x=int(j*src_w/width)
            y=int(i*src_h/height)
            dst[i,j]=img[y,x]
    return dst

def rotate_90(img):
    h,w=img.shape[:2]
    rotate=np.zeros((w,h,3),dtype=np.uint8)
    for i in range(h):
        for j in range(w):
            rotate[j,h-1-i]=img[i,j]
    return rotate

def rotate_90_cc(img):
    h,w=img.shape[:2]
    rotate=np.zeros((w,h,3),dtype=np.uint8)
    for i in range(h):
        for j in range(w):
            rotate[w-1-j,i]=img[i,j]
    return rotate

def rotate_180(img):
    h,w=img.shape[:2]
    rotate=np.zeros((h,w,3),dtype=np.uint8)
    for i in range(h):
        for j in range(w):
            rotate[h-1-i,w-1-j]=img[i,j]
    return rotate

def hori_flip(img):
    h,w=img.shape[:2]
    flip=np.zeros_like(img)
    for i in range(h):
        for j in range(w):
            flip[i,w-1-j]=img[i,j]
    return flip

def vert_flip(img):
    h,w=img.shape[:2]
    flip=np.zeros_like(img)
    for i in range(h):
        for j in range(w):
            flip[h-1-i,j]=img[i,j]
    return flip

resize=resize(img,200,200)
rotate_90=rotate_90(img)
rotate_90_cc=rotate_90_cc(img)
rotate_180=rotate_180(img)
hori_flip=hori_flip(img)
vert_flip=vert_flip(img)

plt.figure(figsize=(12,6))

plt.subplot(331)
plt.imshow(img)
plt.title("Original Image")
plt.axis('off')

plt.subplot(332)
plt.imshow(resize)
plt.title("Original Image")
plt.axis('off')

plt.subplot(333)
plt.imshow(rotate_90)
plt.title("Original Image")
plt.axis('off')

plt.subplot(334)
plt.imshow(rotate_90_cc)
plt.title("Original Image")
plt.axis('off')

plt.subplot(335)
plt.imshow(rotate_180)
plt.title("Original Image")
plt.axis('off')

plt.subplot(336)
plt.imshow(hori_flip)
plt.title("Original Image")
plt.axis('off')

plt.subplot(337)
plt.imshow(vert_flip)
plt.title("Original Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
